```python
import pandas as pd
import requests
import json
import os

# Carrega o batch de dados simulados (sem a coluna "Class")
df = pd.read_csv("data/current_data.csv").drop(columns=["Class"], errors="ignore")

# Prepara dados no formato da API
payload = {
    "inputs": df.to_dict(orient="records")
}

# Chama a API do modelo no MLServer
url = "http://localhost:9000/invocations"
response = requests.post(
    url,
    headers={"Content-Type": "application/json"},
    data=json.dumps(payload)
)

# Verifica resposta
if response.status_code == 200:
    preds = response.json()
    #print(preds)
    df["Class"] = preds['predictions']
    print(df.head())
    df.to_csv("data/predictions_data.csv", index=False)
    print("‚úÖ Predi√ß√µes adicionadas ao predictions_data.csv")
else:
    print("‚ùå Erro ao chamar API:", response.status_code)
    print(response.text)
```
